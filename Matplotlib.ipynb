# Scatter 

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [6,7,9,10,8]

c = ['r','b','g','c','y']  # colours denoted by its first latter
size = [500,300,200,400,100]

plt.xlabel('X-AXIS' , fontsize=20)
plt.ylabel('Y-AXIS' , fontsize=20)

plt.title('Scatter graph')
plt.scatter(a,b, color=c , s=size , marker='*' , label='Mujeeb' , edgecolor='r' , linewidth=1 , linestyle=':' , alpha=0.5)

plt.legend()
plt.show()
```

# - if you want to change the lines colors make it different from others. you can do it by "use_line_collection"

```python
x = [1,2,3,4,5]
y = [2,5,3,6,4]

plt.stem(x,y , linefmt=':' , bottom=2 , markerfmt='r*' , basefmt='g' , label='STEM PLOT' , use_line_collection=False)
# line collection is by default true so I false it

plt.legend() # Necessory for label. label will not work without it
plt.show()
```

## - Now adding arrow to it using annotate

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [2,3,1,5,6]

plt.plot(a,b)

plt.xlabel('Python' , fontsize=20 , style='italic')
plt.ylabel('Popularity' , fontsize=20 , style='italic')

plt.text(3,3 , 'up' ,fontsize=15 , style='italic' , bbox={'facecolor': 'c'})

plt.annotate("Final" , xy=(5,6) , xytext=(2,1) , arrowprops=dict(facecolor="m" , shrink=0))

plt.legend(['Graph'] , fontsize=15 , loc=7 , facecolor='yellow' , edgecolor='red' , shadow=True)

plt.show()
```

```python
a = [1,2,3,4,5]
b = [2,3,1,5,6]

plt.plot(a,b)

plt.xlabel('Python', fontsize=20, style='italic')
plt.ylabel('Popularity', fontsize=20, style='italic')

plt.text(3,3, 'up', fontsize=15, style='italic', bbox={'facecolor': 'c'})

# Assuming you want "Final" at (4, 4)
plt.annotate("Final", xy=(2,1), xytext=(4, 4), arrowprops=dict(facecolor="black", shrink=0.5))

plt.legend(['Graph'], fontsize=15, loc=7, facecolor='yellow', edgecolor='red', shadow=True)

plt.show()
```
