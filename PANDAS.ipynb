```python
# By default ascending is True so just False it
(data2['Team A'].value_counts() + data2['Team B'].value_counts()).sort_values(ascending=False)
```

```python
# Find the most destructive death over batsman of IPL history
# strike rate = (number of runs / number of balls) * 100
# min batsman 200 balls in over 16-200
```

```python
death_over = runs[runs['over'] > 15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsman > 200
all_batsman[x]
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
# Runs score by all these 43 batsmans
# balls played by these 43 batsmans

final = runs[runs['batsman'].isin(batsman_list)]
```

```python
total_runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
total_balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate = (total_runs / total_balls) * 100

strike_rate
```

```python
strike_rate.sort_values(ascending=False).head(5)

# So the following are the top 5 most destructive death over batsmans of the IPL history
```

### - Exploring dropna()

```python
runs.dropna(how='all').shape
# how is automaticcaly any but if you all it means that drop only those rows which dont have anything (means all are null values)
```

```python
runs.dropna(axis=1).shape
# the axis is automaticcly 0 (0 means it will do its work on the basis or rows) and 1 means it will do its work on the basis of columns
```

## - There is another method of "ffill and bfill"
### - ffill will replace the missing value with the upper one value (first fill)
### - similarly bfill will replace the missing value with the lower of that value (back fill)
#### - if 4th row value is 5 and 6th is null and 7th is 7. then if you apply ffill then the 6th one will be 5 and if you apply bfill then the 6th one will be 7

```python
runs['batsman_runs'].fillna(method='ffill')  # this datafram dont have values that why i didnt work but this is the method and similarly with bfill
```
